```python
import pygame
from pygame.locals import *

pygame.init()

X = 600
Y = 600
MARGIN = 10
CELL_X = int(X/6)
CELL_Y = int(Y/6)
Y_SIDE = 4
MAT_SIZE= Y_SIDE*Y_SIDE
ADJUSTER_X = int(X/7)
ADJUSTER_Y = int(Y/7)

BOXX = CELL_X
BOXY = CELL_Y

RED = (255,0,0)
GREEN = (0,255,0)

img_loc_01 = r"C:\Users\Lyndall_Holstein\Desktop\high_lighted_pdfs\workstation_01.jpg"
img_loc_02 = r"C:\Users\Lyndall_Holstein\Desktop\high_lighted_pdfs\masked_worker.jpg"
img_loc_03 = r"C:\Users\Lyndall_Holstein\Desktop\high_lighted_pdfs\dalida_robot.jpg"
img_loc_04 = r"C:\Users\Lyndall_Holstein\Desktop\high_lighted_pdfs\comic_elliot.jpg"

img_01 = pygame.image.load(img_loc_01)
img_01 = pygame.transform.scale(img_01,(X-MARGIN,Y-MARGIN))

img_02 = pygame.image.load(img_loc_02)
img_03 = pygame.image.load(img_loc_03)
img_04 = pygame.image.load(img_loc_04)

img_02 = pygame.transform.scale(img_02,(CELL_X,CELL_Y))
img_03 = pygame.transform.scale(img_03,(CELL_X,CELL_Y))
img_04 = pygame.transform.scale(img_04,(CELL_X,CELL_Y))

D_surf = pygame.display.set_mode((X,Y),0,32)

dict = {}
# count = 0
for i in range(MAT_SIZE):
    dict[i] = {'X':0,'Y':0,'state':False}
    
count = 0        
for i in range(Y_SIDE):
    for j in range(Y_SIDE):
        dict[count]['X'] = (i*CELL_X)+(i*50)
        dict[count]['Y'] = (j*CELL_Y)+(j*50)
        count = count+1  

top_x = dict[0]['X']+13
top_y = dict[0]['Y']+13
bot_x = top_x + BOXX
bot_y = top_y + BOXY
        
while True:
    D_surf.blit(img_01,(10,10))
    for i in range(MAT_SIZE):
        temp_x = dict[i]['X']
        temp_y = dict[i]['Y']
        if dict[i]['state']==False:
            D_surf.blit(img_02,(temp_x+(2*MARGIN),temp_y+(2*MARGIN)))
        if dict[i]['state']==True:
            D_surf.blit(img_03,(temp_x+(2*MARGIN),temp_y+(2*MARGIN)))
    
    pygame.draw.rect(D_surf,GREEN,(top_x,top_y,bot_x,bot_y),4)  
    
    for event in pygame.event.get():
        if event.type == KEYDOWN:
            if event.key == K_w:
                top_y = top_y - (CELL_Y+50)
            if event.key == K_s:
                top_y = top_y + (CELL_Y+50)
            if event.key == K_a:
                top_x = top_x - (CELL_X+50)
            if event.key == K_d:
                top_x = top_x + (CELL_X+50)   
            if event.key == K_q:
                for i in range(16):
                    g = ((dict[i]['X']+13)==top_x)
                    h = ((dict[i]['Y']+13)==top_y)
                    if g and h:
                        dict[i]['state'] = not(dict[i]['state'])                
                
        if event.type == pygame.QUIT:
            pygame.display.quit()
            pygame.quit()
            quit()
   
    pygame.display.update()
```
